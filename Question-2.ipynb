```python
#all libaries I find necessary to complete question
import pandas as pd
import numpy as np
from datetime import datetime, date
import json
```

```python
#use the pandas read_csv function to read the given dataset and use head to display the first 5 records
dat = pd.read_csv('Question 2 Data.csv')
dat.head()
```

```python
#check if any null values in which there is one null record
dat.isnull().sum()
```

```python
#I use the dropna() function to drop all null values from the dataframe
dat = dat.dropna()
```

```python
#check to see if the dropna() function work correctly
dat.isnull().sum()
```

```python
#develop a datetime stamp of the current date
now = pd.to_datetime('now')
now
```

```python
#i add two statements so the second one places the data in similar format to the given data field
dat['DateOfBirth'] = pd.to_datetime(dat['DateOfBirth'])
dat['DateOfBirth'] = pd.to_datetime(dat['DateOfBirth'], format='%d%m%Y')
```

```python
#seems like the income datatype needs to be change to integer
dat.dtypes
```

```python
#Noticed that there was a inbetween the values so removed the space through str.replace and changing the space to nothing
dat['Income'] = dat['Income'].str.replace(' ', '')
#Converting the field to numeric 
dat['Income'] = pd.to_numeric(dat['Income'])
```

```python
#Great an Age field where I get the difference of the year and month of both each current record to our current time, where the month difference has to be less than 0
dat['Age'] = (now.year - dat['DateOfBirth'].dt.year) - ((now.month - dat['DateOfBirth'].dt.month) < 0)
```

```python
#checking to see if the updated filed comes out correctly
dat.head()
```

```python
#Now searching through ages with numpy arrays to determine which age fits in which Age_range
#If fits in a range then is appended to the Age range list as that particular range
Age_range = []
dat_array = dat.to_numpy()
for i in dat_array:
    if i[5] < 18:
        Age_range.append('<18')
    elif i[5] >= 18 and i[5]<=24:
        Age_range.append('18-24')
    elif i[5] >=25 and i[5]<=34:
        Age_range.append('25-34')
    elif i[5] >=35 and i[5] <= 44:
        Age_range.append('35-44')
    elif i[5] >=45 and i[5] <= 54:
        Age_range.append('45-54')
    elif i[5] >=55 and i[5] <= 65:
        Age_range.append('55-65')
    elif i[5] >=65:
        Age_range.append('>65')
        
```

```python
#store the Age_range array into a new field within the dat dataframe called 'AgeRange'
dat['AgeRange'] = Age_range
```

```python
#select the relevant fields for out json file format and select the first 5 records to determine if correct
final_dat = dat[['Name', 'Gender', 'Industry', 'Income', 'AgeRange']]
final_dat.head()
```

```python
#convert the data into a json string using the orient records (as provides the correct )
dat_json_str = final_dat.to_json(orient='records')
#printing the first record to show the current format
dat_json_str[:110]
```

```python
#json loads converts a json string into a dictionary 
dat_json = json.loads(dat_json_str)
#Then json.dumps converts the dictionary into json with indent 2 to make it neat according the the solution of the question
dat_json_formatted = json.dumps(dat_json, indent=2)
print(dat_json_formatted)
```

```python
#writing the solution from above into a json file - where output file name is 'Question2_output.json'
with open('Question2_output.json', 'w', encoding='utf-8') as f:
    json.dump(dat_json, f, indent=2)
```
